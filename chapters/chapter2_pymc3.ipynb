```python
import pymc3 as pm
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

%matplotlib inline
```

## Grid approximation
The receipe
* Define the grid
* Compute the prior value at each parameter value on the grid
* Compute the likelihood at each parameter value
* Multiply the prior and the likelihood to get the unstandardized posterior at each paramter value
* Finally standardize the posterior by dividing each value by sum of all values

```python
num_points = 200

# The experiment that has been conducted on num_tosses yielded num_success
num_successes = 80
num_tosses = 90

# the grid - represents the possible values the probability can take
p_grid = np.linspace(0,1,num_points)

# defining a uniform prior
prior = np.repeat(1,num_points)
prior

# computing value of likelihood at each point in the grid
likelihood = stats.binom.pmf(num_successes, num_tosses, p_grid )

posterior = prior * likelihood
posterior /= sum(posterior)
posterior

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

axes[0].plot(p_grid, prior);
axes[0].set_title('Prior')
axes[0].set_xlabel('Probability of water')
axes[0].set_ylabel('Prior Probability')
axes[1].plot(p_grid, likelihood);
axes[1].set_title('Likelihood')
axes[1].set_xlabel('Probability of water')
axes[2].plot(p_grid, posterior);
axes[2].set_title('Posterior')
axes[2].set_xlabel('Probability of water')
axes[2].set_ylabel('Posterior Probability')

plt.tight_layout()
```
