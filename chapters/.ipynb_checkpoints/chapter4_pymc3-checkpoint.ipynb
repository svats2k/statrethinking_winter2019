### Modeling Random Walk 4.1
* 1000 students, each time we flip a coin they randomly take a step to the left or right.  We need to assess where they would end up after 16 flips

```python
steps = np.random.uniform(-1,1,(1000,16))
steps.shape

# Position after 100 steps
np.sum(steps, axis=1)
```

```python
pm.kdeplot(np.sum(steps, axis=1))
```
